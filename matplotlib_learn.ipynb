### Matplotlib Object Oriented Method

Kegunaan dari OOM: objek ini berguna untuk customized visualisasi

````python
# Create Figure (empty Canvas)
fig = plt.figure()

# Add set of axes to figure
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8]) #left, bottom, width, height (Range 0 - 1)

# Plot on that set of axes
axes.plot(x,y, 'b')
axes.set_xlabel('Set X Label') # Notice the use of set_ to begin methods
axes.set_xlabel('Set Y Label')
axes.set_title('Set Title')
````

### OOM Figure in Figure

Code is a little more complicated, but the advantage is that we now have full control of where the plot axes are placed, and we can easily add more than one axis to the figure:


````python
# Create blank canvas
fig = plt.figure()

axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8]) # main axes
axes2 = fig.add_axes([0.2, 0.5, 0.4, 0.3]) # insert axes

# Larger Figure Axes 1
axes1.plot(x,y, 'b')
axes1.set_xlabel('X_label_axes2')
axes1.set_ylabel('Y_label_axes2')
axes1.set_title('Axes 1 Title')

# Insert Figure Axes 2
axes2.plot(y, x, 'r')
axes2.set_xlabel('X_label_axes2')
axes2.set_ylabel('Y_label_axes2')
axes2.set_title('Axes 2 Title')
````

### Subplots ( )
The plt.subplots() object will act as a more automatic axis manager.
Basic use cases:

````python
# Use similar to plt.figure() except use tuple unpacking to grab fig and axes 
fig, axes = plt.subplots()  # definisi akan banyak objek/gambar

# Now use the axes object to add stuff to plot
axes.plot(x, y, 'r')
axes.set_xlabel('x')
axes.set_ylabel('y')
axes.set_title('title')
````

Then you can specify the number of rows and columns when creating the subplots() object:

````python
# Empty canvas of 1 by 2 subplots
fig, axes = plt.subplots(nrows=1, ncols=2)
````

````python
# Axes in an array of axes to plot on
axes
````

We can iterate through this array:

````python
for ax in axes:
    ax.plot(x, y,'b')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')

# Display the figure object
fig
````

A common issue with matplolib is overlapping subplots or figures. 
We can usefig.tight_layout()orplt.tight_layout()method, which automatically adjusts the positions of the axes on the figure canvas so that there is no overlapping content:

````python
fig, axes = plt.subplots(nrows=2, ncols=1)

for ax in axes:
    ax.plot(x, y,'g')
    ax.set_xlabel('x')
    ax.set_ylabel('y')
    ax.set_title('title')
    
fig
plt.tight_layout()
````

````python
fig, axes = plt.subplots(nrows=1, ncols=3)

ax1 = axes[0]
ax1.plot(y, x,'b')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')

ax2 = axes[1]
ax2.plot(x, y,'g')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('title')

ax3 = axes[2]
ax3.plot(y, y,'r')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_title('title')

plt.tight_layout()  #supaya grafik enak di baca tidak dempet2
````

````python
fig, axes = plt.subplots(nrows=2, ncols=3, figsize=(12,7), dpi=100)

ax1 = axes[0][2] #artinya baris 1 kolom ke 3 
ax1.plot(y, x,'b')
ax1.set_xlabel('x')
ax1.set_ylabel('y')
ax1.set_title('title')

ax2 = axes[1][1] #artinya baris 2 kolom ke 2
ax2.plot(x, x,'g')
ax2.set_xlabel('x')
ax2.set_ylabel('y')
ax2.set_title('title')

ax3 = axes[1][0]
ax3.plot(x, x**3,'g', label='x**3') # membuat 2 garis dalam 1 figure
ax3.plot(x, y,'r--2', label='x**2')
ax3.set_xlabel('x')
ax3.set_ylabel('y')
ax3.set_title('title')
ax3.legend() # kasih penjelasan si garis

ax4 = axes[0][0]
ax4.plot(x, y,'m-*', label='expenses') # membuat 2 garis dalam 1 figure
ax4.set_xlabel('holidays')
ax4.set_ylabel('expanses')
ax4.set_title('holiday vs expenses')
ax4.legend() # kasih penjelasan si garis

ax5 = axes[1][2]
ax5.plot(x, -y,'r--', label='performance') # membuat 2 garis dalam 1 figure
ax5.set_xlabel('age')
ax5.set_ylabel('performance')
ax5.set_title('age vs performance')
ax5.legend(loc=3) # kasih penjelasan si garis loc= 0-4 bisa di coba

plt.tight_layout()
````

## Figure size, aspect ratio and DPI
Matplotlib allows the aspect ratio, DPI and figure size to be specified when the Figure object is created. You can use the figsize and dpi keyword arguments.
   - figsize is a tuple of the width and height of the figure in inches
   - dpi is the dots-per-inch (pixel per inch).


````python
# For Example:
fig = plt.figure(figsize=(8,4), dpi=100)
````

### Saving Figures

Matplotlib can generate high-quality output in a number formats, including PNG, JPG, EPS, SVG, PGF and PDF.

To save a figure to a file we can use the savefig method in the Figure class:

```javascript
fig.savefig("filename.png")
```

Here we can also optionally specify the DPI and choose between different output formats:

```javascript
fig.savefig("filename.png", dpi=200)
```
